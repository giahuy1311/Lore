```python
import numpy as np
import matplotlib.pyplot as plt

x_vals = np.linspace(0, 10, 400)
y_vals = np.linspace(0, 10, 400)
X, Y = np.meshgrid(x_vals, y_vals)

'''
    x = {'x1': 5.442245882003536, 'x2': 3.176784518786956, 'y': 1}
    r = {'x1': '3.987087< x1 <=5.634565', 'x2': '2.720612< x2 <=4.552723'} --> {'y': 1}
    delta {'x1': '>6.161021'}
    delta {'x2': '<=2.195026'}
'''

region_original = ((Y >= (X-5)**2 + 2) & (Y <= -X + 10))

region_lore = ( (3.987087< X) & ( X <=5.634565) & (2.720612< Y) & (Y <=4.552723))

region_counterfactuals = (( X >6.161021) | ( Y <=2.195026))

plt.imshow(region_original.astype(int), extent=(X.min(), X.max(), Y.min(), Y.max()), 
           origin="lower", cmap="Blues", alpha=0.3)

plt.imshow(region_lore.astype(int), extent=(X.min(), X.max(), Y.min(), Y.max()), 
           origin="lower", cmap="Blues", alpha=0.6)

plt.imshow(region_counterfactuals.astype(int), extent=(X.min(), X.max(), Y.min(), Y.max()),
              origin="lower", cmap="Reds", alpha=0.3)

x_curve = np.linspace(0, 10, 400)
plt.plot(x_curve, (x_curve-5)**2 + 2, 'r-', label=r'$x_2 = (x_1 - 5)^2 + 2$')
plt.plot(x_curve, -x_curve + 10, 'b-', label=r'$x_2 = -x_1 + 10$')
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Xác định miền giá trị của x1 và x2
x1_range = np.linspace(4, 7, 300)
x2_range = np.linspace(2, 4, 300)
x1, x2 = np.meshgrid(x1_range, x2_range)

# Xác định vùng thoả mãn điều kiện
feasible_region = (5.36084 < x1) & (x1 <= 5.850503) & (2.57965 < x2) & (x2 <= 3.438712)

# Vẽ đồ thị
plt.figure(figsize=(7, 6))
plt.imshow(feasible_region.astype(int), extent=(x1_range.min(), x1_range.max(), x2_range.min(), x2_range.max()), 
           origin="lower", cmap="Blues", alpha=0.6)

# Vẽ đường biên của miền nghiệm
plt.axvline(5.36084, color="r", linestyle="--", label="$x_1 = 5.36084$")
plt.axvline(5.850503, color="r", linestyle="--", label="$x_1 = 5.850503$")
plt.axhline(2.57965, color="g", linestyle="--", label="$x_2 = 2.57965$")
plt.axhline(3.438712, color="g", linestyle="--", label="$x_2 = 3.438712$")

# Định dạng đồ thị
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.title("Miền nghiệm của bất đẳng thức")
plt.legend()
plt.grid(True)
plt.show()
```
